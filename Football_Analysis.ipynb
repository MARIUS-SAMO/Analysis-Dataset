# Football Data Analysis Project - Python

```python
# Import the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset

EPL = pd.read_csv("EPL_20_21.csv", sep= ";")
EPL.head()
```

```python
EPL.info()
```

```python
EPL.describe()
```

```python
EPL.isna().sum()
```

```python
#create 2 new columns
EPL["MinsPerMatch"] = (EPL["Mins"]/EPL["Matches"]).astype(int)
EPL["GoalsPerMatch"] = EPL["Goals"]/EPL["Matches"]
EPL.head()
```

```python
# Total Goals
Total_Goals = EPL['Goals'].sum()
print("The total number of goals during the season 2021/2022 in EPL is : ", Total_Goals)
```

```python
# Penalty Goals
Total_Penalty = EPL['Penalty_Goals'].sum()
print("The total number of penalty goals during the season 2021/2022 in EPL is : ", Total_Penalty)
```

```python
# Penalty Attempts
Total_Penalty_Attempts = EPL['Penalty_Attempted'].sum()
print("The total number of penalty attempted during the season 2021/2022 in EPL is : ", Total_Penalty_Attempts)
```

```python
# Pie chart for penalties missed vs scored
plt.figure(figsize=(13,7))
not_score_penalty = Total_Penalty_Attempts - Total_Penalty
data = [Total_Penalty, not_score_penalty ]
names = ["Penalties Scored", "Penalty missed"]
color = sns.color_palette("tab10")
plt.pie(data, labels = names, colors = color, autopct = "%.0f%%")
plt.show()
```
